```python
alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
list(enumerate(alphabets))
```

```python
for i, c in enumerate(alphabets):
    print(i, c)
```

```python
int_to_char = dict((i, c) for i, c in enumerate(alphabets))
```

```python
int_to_char
```

```python
char_to_int = dict((c, i) for i, c in enumerate(alphabets))
```

```python
char_to_int
```

```python
seq_length = 1
x = []
y = []

for i in range(len(alphabets)- 1):
    seq_in = alphabets[i:i+seq_length]
    seq_out = alphabets[i+seq_length]
    print(seq_in, '->', seq_out)
    x.append([char_to_int[char] for char in seq_in])
    y.append(char_to_int[seq_out])
```

```python
x
```

```python
y
```

```python
import numpy as np
```

```python
X = np.reshape(x, (len(x), seq_length, 1))
```

```python
X.shape
```

```python
X
```

```python
a = np.array([7,3,0,2,4,8,5])
```

```python
a / 8
```

```python
X = X / 25
```

```python
X
```

```python
a = np.array([3,0,2,3,3,1,2,3,1,0,2,0,0,0,1,2,3])
```

```python
from keras.utils import to_categorical
```

```python
to_categorical(a)
```

```python
y = to_categorical(y)
```

##### Build the model

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
# create the object of RNN
model = Sequential()

# add the LSTM layer
layer1 = LSTM(32, input_shape = (X.shape[1], X.shape[2]))
model.add(layer1)

# add the output layer
layer2 = Dense(26, activation='softmax')
model.add(layer2)
```

```python
model.summary()
```

```python
from keras.utils import plot_model
```

```python
plot_model(model, show_layer_activations=True, show_layer_names=True, show_shapes=True)
```

##### Compile the model

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

##### Train the model

```python
model.fit(X, y, epochs=1000, batch_size=1)
```

##### Evaluate the model

```python
model.evaluate(X, y, batch_size=1)
```

##### Prediction

```python
new = 'E'
new = char_to_int[new]
new = np.reshape(new, (1,1,1))
new = new / 25
# -------------------------------------
result = model.predict(new, verbose=False)
result = np.argmax(result)
int_to_char[result]
```
